### <center> Градиентный спуск

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Определим функцию потерь
def func(w1, w2):
    return 2 * w1 ** 2 + 5 * w2 ** 2

# Найдем производную по первой переменной
def partial_1(w1):
    return 2 * w1
 
# Найдем производную по второй переменной
def partial_2(w2):
      return 2 * w2
```

```python
# Установим первоначальные значения переменных
w1, w2 = 3, 4
 
# Установим количество итераций
iter = 100
 
# Установим скорость обучения
learning_speed = 0.05
```

```python
# Создадим список для последующих значений переменных и ошибки
w1_list, w2_list, e_list = [], [], []
```

```python
for i in range(iter):
 
  # Добавляем значения переменных на каждой итерации в списки
    w1_list.append(w1)
    w2_list.append(w2)
 
  # Рассчитываем уровень ошибки для каждой итерации и добавляем в список
    e_list.append(func(w1, w2))
 
  # Рассчитываем значения частных произодных на каждой итерации
    par_1 = partial_1(w1)
    par_2 = partial_2(w2)
 
  # Обновляем значения в направлении обратном направлению градиента, умноженному на скорость обучения
    w1 = w1 - learning_speed * par_1
    w2 = w2 - learning_speed * par_2
 
    # Выведем итоговые значения модели и значение функции потерь
w1, w2, func(w1, w2)
```

```python
fig = plt.figure(figsize = (14,12))
 
w1 = np.linspace(-5, 5, 1000)
w2 = np.linspace(-5, 5, 1000)
 
w1, w2 = np.meshgrid(w1, w2)
 
f = 2 * w1 ** 2 + 5 * w2 ** 2
 
ax = fig.add_subplot(projection = '3d')
 
ax.plot_surface(w1, w2, f, alpha = 0.4, cmap = 'Blues')
 
ax.text(3, 3.5, 28, 'A', size = 25)
ax.text(0, -0.4, 4, 'B', size = 25)
 
ax.set_xlabel('w1')
ax.set_ylabel('w2')
ax.set_zlabel('f(w1, w2)')
 
# Весь путь алгоритма
ax.plot(w1_list, w2_list, e_list, '.-', c = 'red')
 
plt.show()
```
